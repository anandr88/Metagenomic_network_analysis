code for circos Plot

```python
!pip install nxviz==0.6.3
```

```python
import networkx as nx
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

G1 = nx.from_pandas_edgelist(new_dfV1, 'row', 'col')

figure(figsize=(10, 7))
nx.draw_shell(G1, with_labels=True)
plt.show()
```

```python

for v in G1:
    G1.nodes[v]["class"] = G1.degree(v)

weights = list(new_dfV1['V1'])
```

```python
from matplotlib import font_manager
from nxviz.plots import CircosPlot

c = CircosPlot(G1,figsize=(7, 10),
               node_labels=True,
               edge_width=weights,
               node_grouping="class",
               node_color="class",fontsize = 7)
c.draw()
# plt.savefig("myImage.png", format="png", dpi=300)
plt.show()
```

```python
c = CircosPlot(graph=G1,figsize=(10, 8),node_grouping="class", node_color="class",
               edge_width=weights,node_labels=True,fontsize = 8,node_label_layout="rotation")
c.draw()
c.figure.tight_layout()
plt.savefig("moderateV1.jpg", format="jpg", dpi=300)
plt.show()

```
